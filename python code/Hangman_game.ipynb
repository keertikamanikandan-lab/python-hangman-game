# Hangman game

Hangman is a guessing game where the player tries to guess a hidden word by suggesting letters.

To start, the player is given a word where all letters are replaced with underscores (`_`) so the only information available is the number of letters in the answer.

Each round the player guesses a letter and if the letter guesed is in the answer, its position (or positions) in the word is revealed. Once they successfully fill in the hidden word with its correct letters, the player wins!

If the letter guessed is not in the answer, the player accumulates mistakes. If they make it to six mistakes without finding the answer, the player loses!

The name of the game comes from a drawing of a hanged stick figure. After each mistake, a new element of the hanged figure is drawn.

If the player doesn't guess the word before the last element of the figure is drawn, the player is hanged!

```python
import random

width = 150

# -------------------------------
# RANDOM WORD SELECTION
# -------------------------------
word_list = ["MONSTER", "PYTHON", "HONEY", "APPLE", "ROCKET"]
answer_word = random.choice(word_list).upper()

# -------------------------------
# GAME SETUP
# -------------------------------
hidden_word = "_ " * len(answer_word)

print("WELCOME TO HANGMAN GAME !\n".center(width))
print("Guess the WORD before the man is fully HANGED!\n".center(width))

print("   -----".center(width))
print("   |   |".center(width))
print("       |".center(width))
print("       |".center(width))
print("       |".center(width))
print("       |".center(width))

print("===================\n".center(width))
print("YOUR GAME STARTS NOW !!!!!!!!\n".center(width))

incorrect_letters = []
correct_letters = []
remaining_mistakes = 6

# -------------------------------
# ASCII HANGMAN STAGES
# -------------------------------
hangman_stages = [
    """
     ------
     |    |
     O    |
   / | \\  |
    / \\   |
          |
     ============
     """,

    """
     ------
     |    |
     O    |
   / | \\  |
    /     |
          |
     ============
     """,

    """
     ------
     |    |
     O    |
   / | \\  |
          |
          |
     ============
     """,

    """
     ------
     |    |
     O    |
   / |    |
          |
          |
     ============
     """,

    """
     ------
     |    |
     O    |
      |    |
          |
          |
     ============
     """,

    """
     ------
     |    |
     O    |
          |
          |
          |
     ============
     """,

    """
     ------
     |    |
          |
          |
          |
          |
     ============
     """
]

print(f"Hidden word is : {hidden_word}")

# -------------------------------
# MAIN GAME LOOP
# -------------------------------
while remaining_mistakes > 0:

    # INPUT VALIDATION
    while True:
        picked_letter = input("\nGuess the Letter : ").upper()

        if len(picked_letter) == 1 and picked_letter.isalpha():
            break

        print("\nPlease enter exactly ONE LETTER\n")

    # CHECK DUPLICATE GUESS
    if picked_letter in correct_letters or picked_letter in incorrect_letters:
        print("\nAlready Guessed! Please Try another Letter.\n")
        continue

    # CORRECT GUESS
    if answer_word.find(picked_letter) != -1:

        print(f"\nCORRECT! The word contains the letter {picked_letter}!\n")

        correct_letters.append(picked_letter)

        hidden_word = ""

        for i in answer_word:

            if i in correct_letters:
                hidden_word += i + " "

            else:
                hidden_word += "_ "

        print(f"Word : {hidden_word}\n".center(width))

        # WIN CONDITION
        if "_" not in hidden_word:

            print("🎉 CONGRATULATIONS! YOU WIN! 🎉\n".center(width))
            print(f"✔ The word was : {answer_word}\n".center(width))

            break

    # WRONG GUESS
    else:

        incorrect_letters.append(picked_letter)

        remaining_mistakes -= 1

        print(f"\nWRONG! You have only {remaining_mistakes} chances left.\n")

        print(hangman_stages[remaining_mistakes])

# -------------------------------
# GAME OVER
# -------------------------------
if remaining_mistakes == 0:

    print("💀 GAME OVER! YOU LOST! 💀\n".center(width))

    print(f"✔ The correct word was : {answer_word}".center(width))
```
